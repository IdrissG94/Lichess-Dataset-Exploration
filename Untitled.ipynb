```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
```

```python
games = pd.read_csv("games.csv")
del games['created_at']
del games['last_move_at']
del games['increment_code']
```

```python
elo_w = games['white_rating']
elo_b = games['black_rating']
average_rating = np.round((np.mean(elo_w)+np.mean(elo_b))/2,2)
median_rating = (np.median(elo_w)+np.median(elo_b))/2

# CALCUL DU NOMBRE TOTAL DE JOUEURS DIFFERENTS
id_players = np.hstack((np.array(games['black_id']),np.array(games['white_id'])))
id_player = np.unique(id_players)

print("Nombre total de parties : ",games.shape[0])
print("Nombre total de joueurs différents :",id_player.size)

# l'élo n'étant pas fix au court du temps on peut se permettre de prendre plusieurs classements élos du même joueur

print("Moyenne du classement élo des joueurs sur chaque partie :",average_rating)
print("Médiane du classement élo des joueurs sur chaque partie :",median_rating)
print("Classement élo maximum :",np.max([elo_w,elo_b]))
```

```python
# VISUALISATION DES DONNEES SOUS LA FORME DE GRAPHIQUES
elo = np.hstack((elo_w,elo_b))
plt.hist(elo,bins=range(750,2750,100))
plt.title("Répartition des joueurs en fonction du niveau élo")
plt.xlabel("Classement Elo")
plt.ylabel("Nombre de joueurs")
```

```python
# POURCENTAGE DE VICTOIRE EN FONCTION DE LA COULEUR

w_white = games[games['winner'] == 'white'].shape[0]/20058
w_black = games[games['winner'] == 'black'].shape[0]/20058
draw = 1 - (w_white + w_black)
labels = 'blancs','noirs','nul'
fig, ax = plt.subplots()
ax.pie([w_white,w_black,draw], labels=labels, autopct='%1.1f%%',colors=['forestgreen','darkred','lightgrey'])
plt.title("Pourcentage de victoire en fonction de la couleur")
```

```python
# VISUALISATION DU WINRATE EN FONCTION DE LA DIFFERENCE D'ELO
games['diff_elo'] = games['white_rating'] - games['black_rating']
```

```python
adv_blanc = games[games['diff_elo']>100]
adv_noir = games[games['diff_elo']<100]
adv_normal = games[(games['diff_elo']<=100)&(games['diff_elo']>=100)]
w_white1 = adv_blanc[adv_blanc['winner'] == 'white'].shape[0]/adv_blanc.shape[0]
w_black1 = adv_blanc[adv_blanc['winner'] == 'black'].shape[0]/adv_blanc.shape[0]
draw1 = 1 - (w_white1 + w_black1)
w_white2 = adv_noir[adv_noir['winner'] == 'white'].shape[0]/adv_noir.shape[0]
w_black2 = adv_noir[adv_noir['winner'] == 'black'].shape[0]/adv_noir.shape[0]
draw2 = 1 - (w_white2 + w_black2)
w_white3 = adv_normal[adv_normal['winner'] == 'white'].shape[0]/adv_normal.shape[0]
w_black3 = adv_normal[adv_normal['winner'] == 'black'].shape[0]/adv_normal.shape[0]
draw3 = 1 - (w_white3 + w_black3)
```

```python
fig, ax = plt.subplots()
ax.pie([w_white1,w_black1,draw1], labels=labels, autopct='%1.1f%%',colors=['forestgreen','darkred','lightgrey'])
plt.title("Pourcentage de victoire en fonction de la couleur (avantage elo blanc de 100 points)")
```

```python
fig, ax = plt.subplots()
ax.pie([w_white2,w_black2,draw2], labels=labels, autopct='%1.1f%%',colors=['darkred','forestgreen','lightgrey'])
plt.title("Pourcentage de victoire en fonction de la couleur (avantage elo noir de 100 points)")
```
